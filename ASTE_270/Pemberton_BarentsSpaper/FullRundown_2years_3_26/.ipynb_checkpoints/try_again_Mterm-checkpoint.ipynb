# do the gateway transport for one singular period

```python
# Create masks for each gate
mk_dict = {}

for gate, x_gate, y_gate in zip(['BSO', 'SFJ', 'FJNZ'],
                                [x_bso, x_sfj, x_fjnz],
                                [y_bso, y_sfj, y_fjnz]):

    mk = np.zeros((50, len(x_gate)))
    for z, (i, j) in enumerate(zip(y_gate, x_gate)):
        mk[:, z] = hfC[:, i, j]
    mk[~np.isnan(mk)] = 1
    mk_dict[gate] = mk
```

```python
print(heat_transport_timeseries['BSO'].shape)

# select some example year we want
allyears = np.arange(2003,2018,1)
iyear = np.where(allyears == 2007)
imonth = 3  # (snap taken at april 1 for march average)
mygate = 'BSO'
mk = mk_dict[mygate]

# select the values for some example month
adv_th_sel = heat_transport_timeseries[mygate][iyear,imonth][0][0]*mk      # degC.m^3/s, shape depth,length
adv_slt_sel = salt_transport_timeseries[mygate][iyear,imonth][0][0]*mk     # PSU.m^3/s, shape depth,length
binned_theta_sel = binned_theta_timeseries[mygate][iyear,imonth][0][0]*mk  # index, shape depth,length
binned_salt_sel = binned_salt_timeseries[mygate][iyear,imonth][0][0]*mk    # index, shape depth,length
theta_sel = theta_timeseries[mygate][iyear,imonth][0][0]*mk  # index, shape depth,length
salt_sel = salt_timeseries[mygate][iyear,imonth][0][0]*mk    # index, shape depth,length
```

```python
# now we need DF convergence for this column (this + KPP will be our internal transformations)
Mx = np.zeros((nS,nT))
My = np.zeros((nS,nT))

# for each index in length
for i in range(adv_th_sel.shape[1]):
    for k in range(0,50-1):
    # get the index of the point we want to bin
        if ~np.isnan(binned_salt_sel[k,i]):
            iTpt = int(binned_theta_sel[k,i])
            iSpt = int(binned_salt_sel[k,i])
    
            # do the binning, here we need to normalize twice because the J vector is normalized twice
            Mx[iSpt,iTpt] += adv_slt_sel[k,i] / binwidthT[iTpt] / binwidthS[iSpt] * 1e-6  # Sv/degC
            My[iSpt,iTpt] += adv_th_sel[k,i] / binwidthT[iTpt] * 1e-6  # Sv/PSU

# fix this
binwidthStile = np.tile(binwidthS[:,np.newaxis],(1,nT))
binwidthTtile = np.tile(binwidthT[np.newaxis,:],(nS,1))
dMx_dS = -np.gradient(Mx, axis=1) / binwidthStile   # ∂Jx/∂S   Sv/degC/PSU
dMy_dT = -np.gradient(My, axis=0) / binwidthTtile # ∂Jy/∂T   Sv/degC/PSU

dMy_dT[My==0] = 0
dMx_dS[Mx==0] = 0



divM = dMx_dS + dMy_dT  # Final divergence in Sv/°C/PSU
divM[Mx == 0] = 0
```

```python
# Initialize m(S,T) array: shape [nT, nS]
m = np.zeros((nT, nS))  # Sv/°C/PSU

# Loop over all points along the gate
for i in range(adv_th_sel.shape[1]):
    for k in range(0, 50 - 1):  # assuming vertical levels
        if ~np.isnan(binned_salt_sel[k, i]):
            iTpt = int(binned_theta_sel[k, i])  # T index (0 to nT-1)
            iSpt = int(binned_salt_sel[k, i])   # S index (0 to nS-1)

            # Skip out-of-bounds indices just in case
            if 0 <= iTpt < nT and 0 <= iSpt < nS:
                # Estimate volume flux using TH or SLT
                # Option 1: Use SLT (salt flux in PSU·m³/s)
                if theta_sel[k, i] != 0:
                    volume_flux = adv_th_sel[k, i] / theta_sel[k, i]  # m³/s
                    # print(volume_flux)
                    # volume_flux = adv_slt_sel[k, i] / salt_sel[k, i]  # m³/s
                    # print(volume_flux)
                    # print()
                else:
                    continue

                # Option 2: Alternatively use temperature
                # volume_flux = adv_th_sel[k, i] / temp[k, i]

                # Normalize by bin area (°C × PSU), convert to Sv
                bin_area = binwidthT[iTpt] * binwidthS[iSpt]
                m[iSpt, iTpt] += volume_flux / bin_area * 1e-6  # Sv/°C/PSU
```

```python
n_terms = 5
divergences = []

for i in range(n_terms):
    Jx = JtermsS[i]  # Sv/°C
    Jy = JtermsT[i]  # Sv/PSU

    # Compute gradient in S and T directions
    dJx_dS = -np.gradient(Jx, axis=1) / binwidthStile  # ∂Jx/∂S
    dJy_dT = -np.gradient(Jy, axis=0) / binwidthTtile  # ∂Jy/∂T

    # Optionally mask where J = 0 to avoid noise
    dJx_dS[Jx == 0] = 0
    dJy_dT[Jy == 0] = 0

    div = dJx_dS + dJy_dT  # Sv/°C/PSU
    divergences.append(div)

# Compute sum of tendencies (i.e., predicted ∂v/∂t)
sum_divs = sum(divergences)
divergences.append(sum_divs)

# Plot all terms
fig, axs = plt.subplots(2, 3, figsize=(16, 9), constrained_layout=True)

vmin = min(np.min(d) for d in divergences)
vmax = max(np.max(d) for d in divergences)

for i, ax in enumerate(axs.flat):
    pcm = ax.pcolormesh(Sbin,Tbin,divergences[i], cmap='seismic', vmin=-5, vmax=5)
    ax.set_title(titles[i])
    fig.colorbar(pcm, ax=ax, orientation='vertical', label='Sv/°C/PSU')
    ax.set_xlim(33,35.7)
    ax.set_ylim(-2.4,8)

plt.suptitle("Divergence of J Components for T–S Volume Budget", fontsize=16)
plt.show()
```
