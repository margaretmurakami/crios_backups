```python
# gates_mask starts as NaN everywhere
gates_mask = np.full((ny, nx), np.nan, dtype=float)

def _mark_points(mask, xs, ys, code, ny, nx, name="gate"):
    """
    Mark (y, x) points in mask with 'code'.
    If a point already has a different non-NaN code, set it to 3 (overlap).
    Bounds are clipped to the grid silently.
    """
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if xs.shape != ys.shape:
        raise ValueError(f"{name}: x/y length mismatch: {xs.shape} vs {ys.shape}")

    # clip to valid indices just in case
    xi = np.clip(xs.astype(int), 0, nx-1)
    yi = np.clip(ys.astype(int), 0, ny-1)

    for j, i in zip(yi, xi):
        cur = mask[j, i]
        if np.isnan(cur):
            mask[j, i] = code
        elif cur == code or cur == 3:
            # already same code or already overlap — leave as is
            continue
        else:
            mask[j, i] = 3  # overlap with different code
    return mask

# ---- mark H gates with code = 1 ----
gates_mask = _mark_points(gates_mask, x_bsoh, y_bsoh, 1, ny, nx, name="bsoh")
gates_mask = _mark_points(gates_mask, x_spfjh, y_spfjh, 1, ny, nx, name="spfjh")

# ---- mark V gates with code = 2 ----
gates_mask = _mark_points(gates_mask, x_bsov,  y_bsov,  2, ny, nx, name="bsov")
gates_mask = _mark_points(gates_mask, x_spfjv, y_spfjv, 2, ny, nx, name="spfjv")
gates_mask = _mark_points(gates_mask, x_fjnzv, y_fjnzv, 1, ny, nx, name="fjnzv")
gates_mask = _mark_points(gates_mask, x_nzruv, y_nzruv, 1, ny, nx, name="nzruv")

# Optional: if you prefer 0 instead of NaN for “not a gate”
# gates_mask = np.nan_to_num(gates_mask, nan=0.0)
```

```python

def create_TS_mesh(tsstr,nS,nT,npoints, binned_salinity, binned_theta, attr,idxs):
    '''
    Inputs:
        nS: binsSLT_edges.shape[0]-1
        nT: binsTH_edges.shape[0]-1
        binned_salinity: the array of shape nz, ny, nx of the indices of salinity in the salt bins
        binned_theta: same as above but for theta
        attr: the attribute we want to bin, ie advection, diffusion etc.
        idxs: np.where(mymsk == iB) or whatever indices in mskBasin we are looking at

    Outputs:
        returns an nS by nT shaped array with the summed values within the attr (like volume)
    '''
    
    mesh = np.zeros((len(tsstr),nS, nT, npoints))
    tn = 0
    for t in range(len(tsstr)):
        if len(attr.shape) == 4:
            # time x nz x ny x nx
            thisvol = attr[t][:,idxs[0],idxs[1]]
            thissalt = binned_salinity[t][:,idxs[0],idxs[1]]
            thistemp = binned_theta[t][:,idxs[0],idxs[1]]
        elif len(attr.shape) == 3:
            # time x ny x nx
            thisvol = attr[t][idxs[0],idxs[1]]
            thissalt = binned_salinity[t][idxs[0],idxs[1]]
            thistemp = binned_theta[t][idxs[0],idxs[1]]
            
        # trim the nan values
        thisvol = np.where(np.isnan(thisvol), 0, thisvol)
        thissalt = np.where(np.isnan(thissalt), -1, thissalt)  # Replace NaN with -1
        thistemp = np.where(np.isnan(thistemp), -1, thistemp)
        
        # create the mesh
        meshx = np.zeros((nS,nT,npoints))
        
        # create local timed mesh
        np.add.at(meshx, ( thissalt.astype(int), thistemp.astype(int), np.arange(0,npoints,1)), thisvol[...])  # this should work to add at bins
        
        # add to big mesh
        mesh[tn,:,:,:] = meshx
        del meshx

        tn += 1
    return mesh
```

```python

def surface_contrib_JT(zconv_top_heat, swtop, rcp, fill_last=0.0):
    """
    zconv_top_heat: (ny, nx)
    swtop:          (nz, ny, nx)
    rcp:            scalar
    fill_last:      value for bottom slice (k = nz-1), usually 0.0 or np.nan
    returns:
      JsurfT:       (nz, ny, nx)  # Sv / PSU
    """
    nz, ny, nx = swtop.shape

    eT = zconv_top_heat.reshape(1, ny, nx)  # (1,ny,nx) for broadcast

    J = np.empty_like(swtop, dtype=float)

    # k = 0: (eT - fT[1]) / rcp / dT / dS * 1e-6
    J[0] = (eT[0] - swtop[1]) / rcp if np.ndim(binwidthT)==0 else \
           (eT[0] - swtop[1]) / rcp

    # 1 .. nz-2: -(fT[k+1]-fT[k]) / rcp / dT / dS * 1e-6
    J[1:nz-1] = -(swtop[2:nz] - swtop[1:nz-1]) / rcp

    # bottom slice (k = nz-1): no k+1; choose your boundary convention
    J[-1] = fill_last
    return J

Ft_surftestT = surface_contrib_JT(zconv_top_heat,swtop,myparms['rcp'])    # this is in degC.m^3/s
```

```python
def _bincount_sum_with_nan(idx, vals, nout):
    """
    NaN-aware per-bin sum:
    if *all* entries in a bin are NaN, that bin returns NaN;
    otherwise NaNs are ignored and finite values are summed.
    """
    # track counts of non-nan contributions
    finite = np.isfinite(vals)
    sums   = np.bincount(idx[finite], vals[finite], minlength=nout).astype(float)
    counts = np.bincount(idx[finite], None, minlength=nout).astype(float)
    out = sums
    out[counts == 0] = np.nan
    return out

def plot_figure4_components(
    THETA,
    boundsT,
    binmidT,
    binwidthT1,
    *,
    iiT=None,
    mask=None,
    tendency_name="tmptend",
    **terms
):
    """
    Bin arbitrary component terms by temperature using binmidT intervals and plot:
      Top panel:   each term vs. T (edge-based, divided by binwidthT1)
      Bottom panel: sum(all components except 'tendency_name') vs the tendency term

    Parameters
    ----------
    THETA : array-like
        Tracer field used for binning. Shape must match each term.
    boundsT : 1D array-like
        Bin *edges* (length nT+1). x-locations will be boundsT[1:-1] (length nT-1).
    binmidT : 1D array-like
        Mid-bin edges used to define intervals: binmidT[i] <= THETA < binmidT[i+1]
        Length nT.
    binwidthT1 : 1D array-like
        Edge bin widths (length nT-1), used to convert summed fluxes to edge-based G.
    iiT : slice or 1D index array, optional
        Selection along the temperature-edge axis (length nT-1). Defaults to all.
    mask : array-like, optional
        Same shape as THETA (and each term). Elements where mask is falsy are ignored.
        If provided, terms are multiplied by mask and NaNs are set to zero for counting.
    tendency_name : str, default 'tmptend'
        The key in **terms that should be treated as the tendency curve on panel 2.
    **terms : dict of name -> array-like
        Component arrays to include (e.g., tmpkpp=..., ADVh=..., vdiff=..., surface=..., etc.)
        All arrays must have the same shape as THETA.

    Notes
    -----
    - The binning follows the “new” interpretation used previously:
        binmidT[i] <= THETA < binmidT[i+1], for i = 0..(nT-2)
      which yields nT-1 edge bins.
    - Each component is summed per bin, then divided by `binwidthT1` to give an
      edge-based G-like quantity per temperature edge.
    """

    # --- validate & prepare shapes ---
    THETA      = np.asarray(THETA)
    boundsT    = np.asarray(boundsT)
    binmidT    = np.asarray(binmidT)
    binwidthT1 = np.asarray(binwidthT1)

    if mask is not None:
        mask = np.asarray(mask)
        if mask.shape != THETA.shape:
            raise ValueError("mask must have the same shape as THETA")

    nT   = boundsT.size - 1
    nTm1 = nT - 1
    if binmidT.size != nT:
        raise ValueError("binmidT must have length nT (= len(boundsT)-1)")
    if binwidthT1.size != nTm1:
        raise ValueError("binwidthT1 must have length nT-1 (= len(boundsT)-2 + 1)")

    # x-locations for edge-based curves (length nT-1)
    x = boundsT[1:-1]

    # default selection
    if iiT is None:
        iiT = slice(None)

    # --- flatten for binning ---
    theta_flat = THETA.ravel()

    # build the bin index via binmidT intervals
    # binmidT[i] <= THETA < binmidT[i+1], i = 0..nT-2
    bin_idx_mid = np.digitize(theta_flat, binmidT, right=False) - 1
    valid_mid   = (bin_idx_mid >= 0) & (bin_idx_mid < nTm1) & np.isfinite(theta_flat)

    # if mask provided, fold it into validity
    if mask is not None:
        valid_mid &= (mask.ravel().astype(bool))

    idx_mid = bin_idx_mid[valid_mid]

    # --- per-term binning and edge conversion ---
    G_terms = {}  # name -> (nT-1,) array
    for name, arr in terms.items():
        arr = np.asarray(arr)
        if arr.shape != THETA.shape:
            raise ValueError(f"term '{name}' must have the same shape as THETA")

        vals = arr.ravel().astype(float)

        # Apply mask behavior similar to your snippets:
        #   multiply by mask, set NaNs to 0 for accumulation, but preserve 'all-NaN' as NaN via helper.
        if mask is not None:
            vals = vals * mask.ravel()

        # NaN-aware bin sums (helper will return NaN for bins with no finite contributions)
        S = _bincount_sum_with_nan(idx_mid, vals[valid_mid], nTm1)

        # Convert to edge-based G by dividing by binwidthT1
        G_terms[name] = S / binwidthT1

    # Separate tendency from components-to-sum
    tend_curve = G_terms.get(tendency_name, None)
    comp_names = [k for k in G_terms.keys() if k != tendency_name]

    # Sum across components (excluding tendency)
    if comp_names:
        G_sum = np.nansum(np.vstack([G_terms[k] for k in comp_names]), axis=0)
    else:
        G_sum = np.full(nTm1, np.nan)

    # --- plot ---
    plt.figure(figsize=(14, 4))

    # Top: every component vs T
    ax1 = plt.subplot(1, 1, 1)
    for name in comp_names:
        ax1.plot(x[iiT], G_terms[name][iiT], linewidth=2, label=name)
    if not comp_names:
        ax1.plot([], [], label="(no components provided)")
    ax1.grid(True)
    ax1.set_xlabel('T')
    ax1.set_ylabel('G_T [m^3/s]')
    ax1.set_title('Components by T (edge-based)')
    ax1.legend(ncol=2)
    ax1.axis('tight')

    plt.tight_layout()

    return G_terms
```

```python

def _bincount2d_sum_with_nan(iT, iS, vals, nTm1, nSm1):
    """
    NaN-aware per-(S,T) bin sum on integer bin indices.
    Returns (nSm1, nTm1) with NaN where a bin had zero finite contributions.
    """
    # Flatten 2D indices to 1D bins
    flat = iS.astype(np.int64) * nTm1 + iT.astype(np.int64)
    finite = np.isfinite(vals)
    size = nTm1 * nSm1

    sums   = np.bincount(flat[finite], vals[finite], minlength=size).astype(float)
    counts = np.bincount(flat[finite], None,        minlength=size).astype(float)

    out = sums.reshape(nSm1, nTm1)
    out[counts.reshape(nSm1, nTm1) == 0] = np.nan
    return out

def accumulate_J_terms_TS(
    THETA, SALT,
    binmidT, binmidS,
    binwidthT1, binwidthS1,
    *,
    mask=None,
    scale=1e-6,
    **terms
):
    """
    Bin arbitrary component terms onto (S,T) *edge* bins defined by binmid arrays.

      - Bins: binmidT[i] <= THETA < binmidT[i+1],  i=0..nT-2  (nT-1 T-edge bins)
              binmidS[j] <= SALT  < binmidS[j+1],  j=0..nS-2  (nS-1 S-edge bins)
      - For each term: sum values per (j,i) bin, then divide by (binwidthS1[j] * binwidthT1[i]).
      - NaN semantics: if a bin has zero finite contributions → NaN; otherwise NaNs are ignored.

    Returns dict: name -> array (nS-1, nT-1) in (S,T) order.
    """
    THETA   = np.asarray(THETA)
    SALT    = np.asarray(SALT)
    binmidT = np.asarray(binmidT)
    binmidS = np.asarray(binmidS)
    binwidthT1 = np.asarray(binwidthT1)
    binwidthS1 = np.asarray(binwidthS1)

    if THETA.shape != SALT.shape:
        raise ValueError("THETA and SALT must have the same shape")
    if mask is not None:
        mask = np.asarray(mask)
        if mask.shape != THETA.shape:
            raise ValueError("mask must have the same shape as THETA/SALT")

    nT = binmidT.size
    nS = binmidS.size
    nTm1 = nT - 1
    nSm1 = nS - 1
    if nTm1 <= 0 or nSm1 <= 0:
        raise ValueError("binmidT and binmidS must each have length >= 2")
    if binwidthT1.size != nTm1 or binwidthS1.size != nSm1:
        raise ValueError("binwidthT1 must have len nT-1 and binwidthS1 must have len nS-1")

    # Build (edge) bin indices via the mid-edge intervals
    th = THETA.ravel()
    sa = SALT.ravel()
    iT = np.digitize(th, binmidT, right=False) - 1
    iS = np.digitize(sa, binmidS, right=False) - 1

    valid = (
        (iT >= 0) & (iT < nTm1) &
        (iS >= 0) & (iS < nSm1) &
        np.isfinite(th) & np.isfinite(sa)
    )
    if mask is not None:
        valid &= mask.ravel().astype(bool)

    iT = iT[valid]
    iS = iS[valid]

    # Precompute 2D “cell area” in T–S space for post-division
    denom = np.outer(binwidthS1, binwidthT1)  # (nS-1, nT-1)

    out = {}
    for name, arr in terms.items():
        arr = np.asarray(arr)
        if arr.shape != THETA.shape:
            raise ValueError(f"term '{name}' must have shape {THETA.shape}")
        vals = arr.ravel()[valid].astype(float)

        # NaN-aware sum per (S,T) bin
        summed = _bincount2d_sum_with_nan(iT, iS, vals, nTm1, nSm1)

        # Convert to J (per-edge in T and S): divide by (ΔT * ΔS)
        J = summed / denom

        # Optional scaling (e.g., to Sv/PSU/°C)
        if scale is not None:
            J = J * scale

        out[name] = J

    return out

# ---- Example wiring to match your variable names & output shape ----
# J is a dict of (nS-1, nT-1) arrays
J = accumulate_J_terms_TS(
    THETA, SALT,
    binmidT.flatten(), binmidS.flatten(),
    binwidthT1.flatten(), binwidthS1.flatten(),
    mask=mymsk3d,       # or mask=None
    scale=1e-6,                # your original scaling
    ADVhT=ADVhT*mymsk3d, ADVrT=ADVrT*mymsk3d, DFhT=DFhT*mymsk3d, DFrT=DFrT*mymsk3d, Ft_surftestT=Ft_surftestT*mymsk3d, tmpkppT=tmpkppT*mymsk3d
)

# If you still want a stacked array shaped (6, nS-1, nT-1) in a specific order:
order = ["ADVhT","ADVrT","DFhT","DFrT","Ft_surftestT","tmpkppT"]
JtermsT = np.stack([J[k] for k in order], axis=0)

J = accumulate_J_terms_TS(
    THETA, SALT,
    binmidT.flatten(), binmidS.flatten(),
    binwidthT1.flatten(), binwidthS1.flatten(),
    mask=mymsk3d,       # or mask=None
    scale=1e-6,                # your original scaling
    ADVhS=ADVhS*mymsk3d, ADVrS=ADVrS*mymsk3d, DFhS=DFhS*mymsk3d, DFrS=DFrS*mymsk3d, Ft_surftestS=Ft_surftestS*mymsk3d, tmpkppS=tmpkppS*mymsk3d
)

# If you still want a stacked array shaped (6, nS-1, nT-1) in a specific order:
order = ["ADVhS","ADVrS","DFhS","DFrS","Ft_surftestS","tmpkppS"]
JtermsS = np.stack([J[k] for k in order], axis=0)
```

```python
# maybe we should check if in fact the JtermsT is the same as the Gterm we validated

import numpy as np

# ---------- utilities ----------
def _digitize_interval(x, edges, *, right=False):
    # index k for edges[k] <= x < edges[k+1]
    return np.digitize(x, edges, right=right) - 1

def _bincount2d_sum_with_nan(i0, i1, vals, n0, n1):
    flat = i0.astype(np.int64) * n1 + i1.astype(np.int64)
    finite = np.isfinite(vals)
    size = n0 * n1
    sums   = np.bincount(flat[finite], vals[finite], minlength=size).astype(float)
    counts = np.bincount(flat[finite], None,        minlength=size).astype(float)
    out = sums.reshape(n0, n1)
    out[counts.reshape(n0, n1) == 0] = np.nan
    return out

# ---------- 1D G_T (matches your "new interpretation") ----------
def GT_from_field_new(THETA, field, binmidT, binwidthT1):
    THETA   = np.asarray(THETA)
    field   = np.asarray(field)
    binmidT = np.asarray(binmidT)
    binwidthT1 = np.asarray(binwidthT1)

    nT   = binmidT.size
    nTm1 = nT - 1
    th   = THETA.ravel()
    fv   = field.ravel()

    iT = _digitize_interval(th, binmidT, right=False)  # mid-edge rule
    valid = (iT >= 0) & (iT < nTm1) & np.isfinite(th)

    S = _bincount2d_sum_with_nan(iT[valid], np.zeros_like(iT[valid]),
                                 fv[valid], nTm1, 1).ravel()  # per-edge sum
    GT = S / binwidthT1  # m^3/s per °C-bin width  -> m^3/s on T-edges
    return GT

# ---------- 2D: build faces and gradJ from weighted inputs ----------
def accumulate_TS_weighted_faces(
    THETA, SALT,
    *,
    boundsT, boundsS,          # cell edges (len nT, nS)
    binmidT, binmidS,          # mid-edges (len nT, nS) for the "new" rule
    binwidthT1, binwidthS1,    # ΔT (len nT-1), ΔS (len nS-1)
    mask=None,
    scale=1e-6,                # m^3/s -> Sv
    terms_T=None,              # dict of T-directed arrays in degC·m^3/s
    terms_S=None               # dict of S-directed arrays in PSU·m^3/s
):
    """
    Returns:
      JT_faces : dict[name] -> (nT,   nS-1)   Sv/PSU  (T-first)
      JS_faces : dict[name] -> (nT-1, nS)     Sv/°C   (T-first)
      gradJ    : (nT-1, nS-1) Sv/(°C·PSU)     (T-first) using divergence of total faces
      J_center : dict[name] -> (nS-1, nT-1)   Sv/(°C·PSU) (optional centered densities)
    """
    THETA = np.asarray(THETA); SALT = np.asarray(SALT)
    if THETA.shape != SALT.shape:
        raise ValueError("THETA and SALT must have the same shape")

    boundsT = np.asarray(boundsT); boundsS = np.asarray(boundsS)
    binmidT = np.asarray(binmidT); binmidS = np.asarray(binmidS)
    dT      = np.asarray(binwidthT1); dS      = np.asarray(binwidthS1)
    nT, nS  = boundsT.size, boundsS.size
    print(binmidT.size,nT,dT.size)
    if any([
        binmidT.size != nT, binmidS.size != nS,
        dT.size != nT-1,    dS.size != nS-1
    ]):
        raise ValueError("Inconsistent T/S grids")

    # flatten + combined validity
    th = THETA.ravel(); sa = SALT.ravel()
    if mask is not None:
        m = np.asarray(mask)
        if m.shape != THETA.shape:
            raise ValueError("mask must match THETA/SALT shape")
        valid0 = m.ravel().astype(bool)
    else:
        valid0 = np.ones(THETA.size, bool)

    # indices for T-faces: T by mid-edges (internal faces), S by centers
    iT_face = _digitize_interval(th, binmidT, right=False)     # 0..nT-2 for internal faces
    iS_cent = _digitize_interval(sa, boundsS,  right=False)    # 0..nS-2 centers
    valid_T = (iT_face >= 0) & (iT_face < nT-1) & (iS_cent >= 0) & (iS_cent < nS-1) \
              & np.isfinite(th) & np.isfinite(sa) & valid0
    iT_face = iT_face[valid_T]; iS_cent_T = iS_cent[valid_T]

    # indices for S-faces: S by mid-edges (internal faces), T by centers
    iS_face = _digitize_interval(sa, binmidS, right=False)     # 0..nS-2
    iT_cent = _digitize_interval(th, boundsT,  right=False)    # 0..nT-2
    valid_S = (iS_face >= 0) & (iS_face < nS-1) & (iT_cent >= 0) & (iT_cent < nT-1) \
              & np.isfinite(th) & np.isfinite(sa) & valid0
    iS_face = iS_face[valid_S]; iT_cent_S = iT_cent[valid_S]

    # ---------- T-faces from degC·m^3/s ----------
    JT_faces = {}
    if terms_T:
        for name, A in terms_T.items():
            vals = np.asarray(A).ravel()[valid_T].astype(float)
            # sum onto (S-center, T-internal-face): shape (nS-1, nT-1) in S-first layout
            summed = _bincount2d_sum_with_nan(iS_cent_T, iT_face, vals, nS-1, nT-1)
            # remove °C, then normalize per-PSU, then scale to Sv:
            JT_int = (summed / dT[None, :]) / dS[:, None] * scale   # Sv/PSU on internal T-faces
            # pad T boundaries with zeros, then transpose to T-first
            JT_pad = np.zeros((nS-1, nT), float)
            JT_pad[:, 1:-1] = JT_int
            JT_faces[name] = JT_pad.T   # (nT, nS-1)

    # ---------- S-faces from PSU·m^3/s ----------
    JS_faces = {}
    if terms_S:
        for name, A in terms_S.items():
            vals = np.asarray(A).ravel()[valid_S].astype(float)
            # sum onto (T-center, S-internal-face) but T-first layout directly:
            summed = _bincount2d_sum_with_nan(iT_cent_S, iS_face, vals, nT-1, nS-1)  # (nT-1, nS-1)
            # remove PSU, then normalize per-°C, then scale to Sv:
            JS_int = (summed / dS[None, :]) / dT[:, None] * scale   # Sv/°C on internal S-faces
            # pad S boundaries with zeros
            JS_pad = np.zeros((nT-1, nS), float)
            JS_pad[:, 1:-1] = JS_int
            JS_faces[name] = JS_pad       # (nT-1, nS)

    # ---------- divergence on centers (T-first, Sv/(°C·PSU)) ----------
    if JT_faces:
        JT_tot = np.nansum(np.stack(list(JT_faces.values()), axis=0), axis=0)  # (nT,   nS-1)
    else:
        JT_tot = np.zeros((nT, nS-1), float)
    if JS_faces:
        JS_tot = np.nansum(np.stack(list(JS_faces.values()), axis=0), axis=0)  # (nT-1, nS)
    else:
        JS_tot = np.zeros((nT-1, nS), float)

    dJT_dT = (JT_tot[1:, :] - JT_tot[:-1, :]) / dT[:, None]   # (nT-1, nS-1)
    dJS_dS = (JS_tot[:, 1:] - JS_tot[:, :-1]) / dS[None, :]   # (nT-1, nS-1)
    gradJ  = dJT_dT + dJS_dS

    # ---------- optional centered densities (Sv/(°C·PSU)) ----------
    J_center = {}
    if terms_T:
        for name, A in terms_T.items():
            vals = np.asarray(A).ravel()[valid_T].astype(float)
            # reuse sums at (S-center, T-internal-face)
            summed = _bincount2d_sum_with_nan(iS_cent_T, iT_face, vals, nS-1, nT-1)
            J_center[name] = (summed / (dS[:, None] * dT[None, :])) * scale
    if terms_S:
        for name, A in terms_S.items():
            vals = np.asarray(A).ravel()[valid_S].astype(float)
            # sums at (T-center, S-internal-face) -> transpose to S-first before dividing
            summed_Tfirst = _bincount2d_sum_with_nan(iT_cent_S, iS_face, vals, nT-1, nS-1)  # (nT-1, nS-1)
            summed = summed_Tfirst.T  # (nS-1, nT-1) S-first
            J_center[name] = J_center.get(name, 0.0) + (summed / (dS[:, None] * dT[None, :])) * scale

    return JT_faces, JS_faces, gradJ, J_center
```

```python
# ---------- helpers ----------
def _bincount2d_sum_with_nan(iT, iS, vals, nTm1, nSm1):
    """
    Sum vals by integer bin indices (iT=0..nTm1-1, iS=0..nSm1-1).
    Returns (nS-1, nT-1) with NaN where a bin received no finite contributions.
    """
    flat = iS.astype(np.int64) * nTm1 + iT.astype(np.int64)
    finite = np.isfinite(vals)
    size = nTm1 * nSm1

    sums   = np.bincount(flat[finite], vals[finite], minlength=size).astype(float)
    counts = np.bincount(flat[finite], None,        minlength=size).astype(float)

    out = sums.reshape(nSm1, nTm1)
    out[counts.reshape(nSm1, nTm1) == 0] = np.nan
    return out


def _build_TS_indices(THETA, SALT, binmidT, binmidS, mask=None):
    """
    Build edge-bin indices using the mid-edge convention:
      binmid[i] <= X < binmid[i+1], i=0..n-2.
    Returns iT, iS filtered to valid+mask and a boolean 'valid' mask for ravel().
    """
    THETA = np.asarray(THETA); SALT = np.asarray(SALT)
    binmidT = np.asarray(binmidT); binmidS = np.asarray(binmidS)

    nT = binmidT.size; nS = binmidS.size
    nTm1, nSm1 = nT - 1, nS - 1
    if nTm1 <= 0 or nSm1 <= 0:
        raise ValueError("binmidT/binmidS must each have length >= 2")

    th = THETA.ravel(); sa = SALT.ravel()
    iT = np.digitize(th, binmidT, right=False) - 1
    iS = np.digitize(sa, binmidS, right=False) - 1

    valid = (
        (iT >= 0) & (iT < nTm1) &
        (iS >= 0) & (iS < nSm1) &
        np.isfinite(th) & np.isfinite(sa)
    )
    if mask is not None:
        m = np.asarray(mask)
        if m.shape != THETA.shape:
            raise ValueError("mask must match THETA/SALT shape")
        valid &= m.ravel().astype(bool)

    return iT[valid], iS[valid], valid, nTm1, nSm1


def _faces_from_centered_Tfirst(JTc, JSc, bc_T="zero", bc_S="zero"):
    """
    Inputs: centered densities, S-first layout  (nS-1, nT-1).
      JTc ~ T-directed density at centers; JSc ~ S-directed density at centers.
    Output: T-first faces
      JT_Tfirst: (nT,   nS-1)  (T-faces)
      JS_Tfirst: (nT-1, nS)    (S-faces)
    """
    nSm1, nTm1 = JTc.shape

    # T-faces on S-first: (nS-1, nT-1+1) = (nS-1, nT)
    JT_faces = np.zeros((nSm1, nTm1 + 1), dtype=float)
    JT_faces[:, 1:-1] = 0.5 * (JTc[:, 1:] + JTc[:, :-1])
    if bc_T == "zero":
        JT_faces[:, 0]  = 0.0;  JT_faces[:, -1] = 0.0
    elif bc_T == "copy":
        JT_faces[:, 0]  = JTc[:, 0];  JT_faces[:, -1] = JTc[:, -1]
    elif bc_T == "periodic":
        JT_faces[:, 0]  = 0.5 * (JTc[:, 0] + JTc[:, -1]);  JT_faces[:, -1] = JT_faces[:, 0]
    else:
        raise ValueError("bc_T must be 'zero', 'copy', or 'periodic'")

    # S-faces on S-first: (nS-1+1, nT-1) = (nS, nT-1)
    JS_faces = np.zeros((nSm1 + 1, nTm1), dtype=float)
    JS_faces[1:-1, :] = 0.5 * (JSc[1:, :] + JSc[:-1, :])
    if bc_S == "zero":
        JS_faces[0,  :] = 0.0;  JS_faces[-1, :] = 0.0
    elif bc_S == "copy":
        JS_faces[0,  :] = JSc[0, :];  JS_faces[-1, :] = JSc[-1, :]
    elif bc_S == "periodic":
        JS_faces[0,  :] = 0.5 * (JSc[0, :] + JSc[-1, :]);  JS_faces[-1, :] = JS_faces[0, :]
    else:
        raise ValueError("bc_S must be 'zero', 'copy', or 'periodic'")

    # Return T-first shapes
    JT_Tfirst = JT_faces.T  # (nT,   nS-1)
    JS_Tfirst = JS_faces.T  # (nT-1, nS)
    return JT_Tfirst, JS_Tfirst


def _div_TS_Tfirst(JT_Tf, JS_Tf, binmidT, binmidS):
    """
    Divergence on centers (nT-1, nS-1) from T-first faces.
      JT_Tf: (nT,   nS-1)   (T-faces)
      JS_Tf: (nT-1, nS)     (S-faces)
    Returns: (nT-1, nS-1) in T-first ordering.
    """
    dT = np.asarray(binmidT[1:] - binmidT[:-1])  # (nT-1,)
    dS = np.asarray(binmidS[1:] - binmidS[:-1])  # (nS-1,)

    dJT_dT = (JT_Tf[1:, :] - JT_Tf[:-1, :]) / dT[:, None]     # (nT-1, nS-1)
    dJS_dS = (JS_Tf[:, 1:] - JS_Tf[:, :-1]) / dS[None, :]     # (nT-1, nS-1)
    return dJT_dT + dJS_dS


# ---------- main ----------
def accumulate_J_terms_TS(
    THETA, SALT,
    binmidT, binmidS,
    binwidthT1, binwidthS1,
    *,
    mask=None,
    scale=1e-6,
    return_faces=True,
    bc_T="zero",
    bc_S="zero",
    **terms
):
    """
    Bin component terms onto centered TS bins and (optionally) build face-staggered fields
    and gradJ in one call.

    Inputs
    ------
    THETA, SALT : same shape arrays (3D or N-D), tracer values for binning
    binmidT, binmidS : 1D arrays of mid-edge coordinates (length nT, nS)
    binwidthT1, binwidthS1 : 1D arrays of central widths (length nT-1, nS-1)
    mask : same shape as THETA (boolean or 0/1); ignored entries are dropped
    scale : multiply final densities (e.g., 1e-6 for m^3/s -> Sv)
    return_faces : if True, also return T-first faces and gradJ
    bc_T, bc_S : boundary condition when averaging to faces: 'zero'|'copy'|'periodic'
    **terms : name -> array (same shape as THETA), values in m^3/s

    Returns
    -------
    out : dict with keys:
      'centered' : dict[name] -> (nS-1, nT-1) centered densities (S-first layout)
      'faces'    : dict with 'JT_Tfirst' (nT, nS-1) and 'JS_Tfirst' (nT-1, nS) if return_faces
      'gradJ_Tfirst' : (nT-1, nS-1) divergence density if return_faces
    """
    # --- validate widths ---
    binmidT = np.asarray(binmidT); binmidS = np.asarray(binmidS)
    binwidthT1 = np.asarray(binwidthT1); binwidthS1 = np.asarray(binwidthS1)

    iT, iS, valid, nTm1, nSm1 = _build_TS_indices(THETA, SALT, binmidT, binmidS, mask=mask)

    if binwidthT1.size != nTm1 or binwidthS1.size != nSm1:
        raise ValueError("binwidthT1 len must be nT-1 and binwidthS1 len must be nS-1")

    denom = np.outer(binwidthS1, binwidthT1)  # (nS-1, nT-1)

    centered = {}
    for name, arr in terms.items():
        A = np.asarray(arr)
        if A.shape != np.asarray(THETA).shape:
            raise ValueError(f"term '{name}' must have shape {THETA.shape}")
        vals = A.ravel()[valid].astype(float)
        summed = _bincount2d_sum_with_nan(iT, iS, vals, nTm1, nSm1)  # m^3/s per bin
        J = summed / denom                                          # m^3/s/(°C·PSU)
        if scale is not None:
            J *= scale                                              # Sv/(°C·PSU)
        centered[name] = J                                          # (nS-1, nT-1)

    result = {"centered": centered}

    if return_faces:
        # Sum components first (centered totals), S-first layout
        if len(centered) == 0:
            JTc = JSc = np.zeros((nSm1, nTm1), dtype=float) * np.nan
        else:
            # By convention: T-directed vs S-directed sets.
            # If you passed separate dicts, adapt as needed; here we try to infer by suffix.
            T_like = [k for k in centered if k.endswith("T")]
            S_like = [k for k in centered if k.endswith("S")]

            if T_like:
                JTc = np.nansum([centered[k] for k in T_like], axis=0)
            else:
                JTc = np.zeros((nSm1, nTm1), dtype=float)

            if S_like:
                JSc = np.nansum([centered[k] for k in S_like], axis=0)
            else:
                JSc = np.zeros((nSm1, nTm1), dtype=float)

        JT_Tfirst, JS_Tfirst = _faces_from_centered_Tfirst(JTc, JSc, bc_T=bc_T, bc_S=bc_S)
        gradJ_Tfirst = _div_TS_Tfirst(JT_Tfirst, JS_Tfirst, binmidT, binmidS)

        result["faces"] = {"JT_Tfirst": JT_Tfirst, "JS_Tfirst": JS_Tfirst}
        result["gradJ_Tfirst"] = gradJ_Tfirst  # (nT-1, nS-1)

    return result
```

```python
# Build once; do not pre-mask terms (mask is handled inside)
outT = accumulate_J_terms_TS(
    THETA, SALT,
    binmidT, binmidS, binwidthT1, binwidthS1,
    mask=mymsk3d, scale=1e-6, return_faces=True,
    bc_T="zero", bc_S="zero",
    ADVhT=ADVhT*mymsk3d, ADVrT=ADVrT*mymsk3d, DFhT=DFhT*mymsk3d, DFrT=DFrT*mymsk3d, Ft_surftestT=Ft_surftestT*mymsk3d, tmpkppT=tmpkppT*mymsk3d,
    ADVhS=ADVhS*mymsk3d, ADVrS=ADVrS*mymsk3d, DFhS=DFhS*mymsk3d, DFrS=DFrS*mymsk3d, Ft_surftestS=Ft_surftestS*mymsk3d, tmpkppS=tmpkppS*mymsk3d,
)

# Centered component fields (S-first, each (nS-1, nT-1))
J_centered = outT["centered"]

# T-first faces with your requested shapes:
JT_Tfirst = outT["faces"]["JT_Tfirst"]  # (nT,   nS-1)
JS_Tfirst = outT["faces"]["JS_Tfirst"]  # (nT-1, nS)

# Divergence density on centers (T-first ordering, matches dVdt density if you use same normalization):
gradJ = outT["gradJ_Tfirst"]            # (nT-1, nS-1)
```
