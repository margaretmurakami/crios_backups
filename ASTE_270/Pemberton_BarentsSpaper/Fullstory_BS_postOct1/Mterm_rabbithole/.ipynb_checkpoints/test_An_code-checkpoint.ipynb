```python
def edges_from_centers(Tc):
    Tf = 0.5*(Tc[:-1] + Tc[1:])
    Te = np.empty(Tc.size + 1)
    Te[1:-1] = Tf
    Te[0]  = 2*Tc[0]  - Tf[0]
    Te[-1] = 2*Tc[-1] - Tf[-1]
    return Te

def walin_G_from_bin_totals(T_centers, Q_bin_centers):
    """
    Q_bin_centers: per-bin totals (degC·m^3/s) aligned to centers; length N.
    Returns G on faces (length N): interior faces from widths, ends via simple extrapolation.
    If you specifically want length N-1 (interior faces only), drop the endpoints.
    """
    T_edges = edges_from_centers(T_centers)         # length N+1
    dT_center = T_edges[1:] - T_edges[:-1]          # length N (bin widths around each center)
    # Map center-integrated totals to face-centered G by dividing by local bin width.
    # Conventionally, G = - (bin_total)/ΔT
    G_centers_proxy = - Q_bin_centers / dT_center   # m^3/s on "center bins"
    # If you truly want faces (length N-1), average neighboring center proxies:
    T_faces = 0.5*(T_centers[:-1] + T_centers[1:])
    G_faces = 0.5*(G_centers_proxy[:-1] + G_centers_proxy[1:])
    return T_faces, G_faces

Ttest,Gtest = walin_G_from_bin_totals(binsTH_centers,df_hconv_binned)
Ttestr,Gtestr = walin_G_from_bin_totals(binsTH_centers,df_rconv_binned)
```

```python
# we need to create zconv_top and swtop
dd = mygrid['RF'][:-1]
swfrac = 0.62*np.exp(dd/0.6)+(1-0.62)*np.exp(dd/20)
swfrac[dd < -200] = 0
swtop=mk3D_mod(swfrac,np.zeros((nz,ny,nx)))*mk3D_mod(RAC*oceQsw,np.zeros((nz,ny,nx)))   # J/s

# zconvtop_heat is here
zconv_top_heat = TFLUX * RAC     # W/m^2 * m^2 = J/s
```

```python
# create the per-cell contributions of the surface terms in zyx

def surface_contrib_JT(zconv_top_heat, swtop, rcp, fill_last=0.0):
    """
    zconv_top_heat: (ny, nx)
    swtop:          (nz, ny, nx)
    rcp:            scalar
    fill_last:      value for bottom slice (k = nz-1), usually 0.0 or np.nan
    returns:
      JsurfT:       (nz, ny, nx)  # Sv / PSU
    """
    nz, ny, nx = swtop.shape

    eT = zconv_top_heat.reshape(1, ny, nx)  # (1,ny,nx) for broadcast

    J = np.empty_like(swtop, dtype=float)

    # k = 0: (eT - fT[1]) / rcp / dT / dS * 1e-6
    J[0] = (eT[0] - swtop[1]) / rcp if np.ndim(binwidthT)==0 else \
           (eT[0] - swtop[1]) / rcp

    # 1 .. nz-2: -(fT[k+1]-fT[k]) / rcp / dT / dS * 1e-6
    J[1:nz-1] = -(swtop[2:nz] - swtop[1:nz-1]) / rcp

    # bottom slice (k = nz-1): no k+1; choose your boundary convention
    J[-1] = fill_last
    return J

Ft_surftest = surface_contrib_JT(zconv_top_heat,swtop,myparms['rcp'])    # this is in degC.m^3/s
```

```python
# we can bin the surface term similar to how we did for diffusive
surf_binned = np.zeros(nT)

for T in range(nT):
    # grab the indices where the binned_theta is this value (z,y,x)
    indices = np.where(binned_theta == T)
    
    # take the nansum of these indices and add it to the correct position in the array (degC.m^3/s)
    tmpsum_surf = np.nansum(Ft_surftest[indices[0],indices[1],indices[2]])

    # take the overall sum of this and add it to the array at that value
    surf_binned[T] = tmpsum_surf
```
