```python
def line_orientation(x, y):
    orientations = np.array([])

    # Iterate through the list of points
    for i in range(1, len(x)):
        if x[i] == x[i - 1] and y[i] != y[i - 1]:
            orientations = np.append(orientations,"V") # vertical line add zonal transp
        elif y[i] == y[i - 1] and x[i] != x[i - 1]:
            orientations = np.append(orientations,"H") # horizontal line add meridional transp
        else:
            orientations = np.append(orientations,"V")  # fill na should be for top of vertical line (ie U)

    # the last one will be the same as the second to last orientation so we can keep consistent sizing
    orientations = np.append(orientations,orientations[-1])

    return orientations
```

Above is showing fluxes, we want the tendency terms.
We need to get the dendency terms using divergence. Ie we need to get these in terms of PSU/s and C/s rather than having these as directional fluxes.

The divergence of the vector field is given as a gradient of the terms with respect to dx, dy, dz

```python
# find G_T
# dxg3d,dyg3d,drf3d
```

```python
ADVx_THex = ADVx_THo[m]
ADVy_THex = ADVy_THo[m]
ADVr_THex = get_aste_tracer(ADVr_TH[m],nfx,nfy)

# Compute the gradients
dADVx_TH = np.gradient(ADVx_THex, axis=2)  # gradient along x-axis (east-west)
dADVy_TH = np.gradient(ADVy_THex, axis=1)
dADVr_TH = np.gradient(ADVr_THex, axis=0)

# divide by bin widths
dADVx_dx_TH = dADVx_TH / get_aste_tracer(dxg3d,nfx,nfy)
dADVy_dy_TH = dADVy_TH / get_aste_tracer(dyg3d,nfx,nfy)
dADVr_dz_TH = dADVr_TH / get_aste_tracer(drf3d,nfx,nfy)

# do divergence
div_adv_TH = dADVx_dx_TH + dADVy_dy_TH + dADVr_dz_TH
```

```python
ADVx_SLTex = ADVx_SLTo[m]
ADVy_SLTex = ADVy_SLTo[m]
ADVr_SLTex = get_aste_tracer(ADVr_SLT[m],nfx,nfy)

# Compute the gradients
dADVx_SLT = np.gradient(ADVx_SLTex, axis=2)  # gradient along x-axis (east-west)
dADVy_SLT = np.gradient(ADVy_SLTex, axis=1)
dADVr_SLT = np.gradient(ADVr_SLTex, axis=0)

# divide by bin widths
dADVx_dx_SLT = dADVx_SLT / get_aste_tracer(dxg3d,nfx,nfy)
dADVy_dy_SLT = dADVy_SLT / get_aste_tracer(dyg3d,nfx,nfy)
dADVr_dz_SLT = dADVr_SLT / get_aste_tracer(drf3d,nfx,nfy)

# do divergence
div_adv_SLT = dADVx_dx_SLT + dADVy_dy_SLT + dADVr_dz_SLT
```
