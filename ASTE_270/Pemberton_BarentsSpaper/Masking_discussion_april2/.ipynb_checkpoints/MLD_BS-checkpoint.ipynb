# define the MLD at all points along this transect from the T and S profiles

We will save a "gate" much like we did for the gateway transports

```python
# Get the indices of your 45E gate from y_lon and x_lon

# Define time and shape parameters
num_years = len(allyears)
num_points = len(y_lon)  # number of points along gate
theta_timeseries = np.full((num_years, 12, nz, num_points), np.nan)
salt_timeseries = np.full((num_years, 12, nz, num_points), np.nan)

iy = 0  # year index

for year in allyears:
    print(f"Processing year: {year}")

    # Get time steps for the year
    years = np.array([str(year)])
    times = {year: np.arange(1, 13, 1)}  # All 12 months
    tsstr, datetimes = get_tsteps(times, fnames, dt_aste, startyr, 1, 1)

    for imonth, ts in enumerate(tsstr):
        file_name = "state_3d_set1"
        meta = parsemeta(os.path.join(dirState, file_name + "." + ts + ".meta"))
        fldlist = np.array(meta["fldList"])
        recs = [np.where(fldlist == var)[0][0] for var in ["THETA", "SALT"]]

        read = [int(ts)]
        THETA, _, _ = rdmds(os.path.join(dirState, file_name), read, returnmeta=True, rec=recs[0])
        SALT, _, _ = rdmds(os.path.join(dirState, file_name), read, returnmeta=True, rec=recs[1])

        # Reshape to full grid
        THETA = np.reshape(THETA, (nz, ny, nx))
        SALT = np.reshape(SALT, (nz, ny, nx))

        # Extract values along the gate
        for ipt, (i, j) in enumerate(zip(y_lon, x_lon)):
            theta_timeseries[iy, imonth, :, ipt] = THETA[:, i, j]
            salt_timeseries[iy, imonth, :, ipt] = SALT[:, i, j]

    iy += 1  # next year
```

```python
# get the approximate location of the polar front by T = 3
theta_timeseries_pf = theta_timeseries.reshape((12*15,nz,len(lat_gate)))
theta_timeseries_pf = theta_timeseries_pf[:,0,:]  # only surface
theta_timeseries_pf.shape
```

```python
lat_idx_maxgrad = np.full((theta_timeseries_pf.shape[0],), np.nan)

for t in range(theta_timeseries_pf.shape[0]):
    sst_profile = theta_timeseries_pf[t, :]  # 1D SST along the gate at time t

    # Skip if missing data
    if np.any(np.isnan(sst_profile)):
        continue

    # Compute centered gradient (or use np.gradient for uniform spacing)
    grad_sst = np.gradient(sst_profile)

    # Get the index of max absolute gradient
    maxgrad_idx = np.argmax(np.abs(grad_sst))

    lat_idx_maxgrad[t] = maxgrad_idx
```

# BY DENSITY: now get the MLD as a line for each of these timesteps

This is following the description in the Montegut 2004 paper for temperature in the North Atlantic

```python
MLD_along_gate = np.full((num_years, 12, num_points), np.nan)
depth = -mygrid['RC']  # positive depth

k0 = 0  # ~5 m
k1 = 1  # ~15 m

for iy in range(num_years):
    for imonth in range(12):
        for ipt in range(num_points):
            temp = theta_timeseries[iy, imonth, :, ipt]
            salt = salt_timeseries[iy, imonth, :, ipt]

            if np.any(np.isnan(temp)) or np.any(np.isnan(salt)):
                print(1)
                continue

            # Compute density profile
            rho = gsw.rho(salt, temp, 0)

            # Approximate T10m and S10m by averaging top 2 levels
            T_10m = 0.5 * (temp[k0] + temp[k1])
            S_10m = 0.5 * (salt[k0] + salt[k1])

            # Compute  = rho(T10m, S10m) - rho(T10m - 0.2C, S10m)
            rho_10m = gsw.rho(S_10m, T_10m, 0)
            rho_ref = gsw.rho(S_10m, T_10m - 0.5, 0)
            delta_sigma =  0.03 #rho_10m - rho_ref

            # Use the criterion: rho(z) > rho_10m + 
            mask = rho > (rho_10m + delta_sigma)
            if np.any(mask):
                k_mld = np.argmax(mask)
                #MLD_along_gate[iy, imonth, ipt] = depth[k_mld]

                if k_mld > 0:
                    # Linear interpolation between levels k_mld-1 and k_mld
                    rho1, rho2 = rho[k_mld - 1], rho[k_mld]
                    z1, z2 = depth[k_mld - 1], depth[k_mld]
                    rho_thresh_val = rho_10m + delta_sigma
                
                    # Interpolated MLD depth
                    mld_interp = z1 + (rho_thresh_val - rho1) / (rho2 - rho1) * (z2 - z1)
                    MLD_along_gate[iy, imonth, ipt] = mld_interp
                else:
                    # No interpolation possible, just assign base level
                    # No level met the MLD threshold
                    i, j = y_lon[ipt], x_lon[ipt]
    
                    # Compute local depth as the sum of layer thicknesses weighted by hFacC
                    local_hfac = hFacC[:, i, j]
                    local_depth = np.nansum(local_hfac * mygrid['DRF'].flatten())  # DRF is layer thickness
                    print(local_depth)
    
                    MLD_along_gate[iy, imonth, ipt] = local_depth
```

```python
# set nan to maximum depth

for iy in range(15):
    for imonth in range(12):
        for ipt in range(num_points):
            i, j = y_lon[ipt], x_lon[ipt]
            local_hfac = hfC[:, i, j]
            local_depth = np.nansum(local_hfac * mygrid['DRF'].flatten())
            # print(local_depth)
        
            # Replace NaNs with local depth
            mask_nan = np.isnan(MLD_along_gate[:, :, ipt])
            MLD_along_gate[:, :, ipt][mask_nan] = local_depth
```
