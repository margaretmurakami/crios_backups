```python
def _bincount_sum_with_nan(idx, vals, nout):
    """
    NaN-aware per-bin sum:
    if *all* entries in a bin are NaN, that bin returns NaN;
    otherwise NaNs are ignored and finite values are summed.
    """
    # track counts of non-nan contributions
    finite = np.isfinite(vals)
    sums   = np.bincount(idx[finite], vals[finite], minlength=nout).astype(float)
    counts = np.bincount(idx[finite], None, minlength=nout).astype(float)
    out = sums
    out[counts == 0] = 0
    return out
```

```python

# define a function to turn the gates as 3D into 1D for T and S
def gateway3D(ADV_gates,binmidTracer,nTm1,binwidth):

    '''
    inputs
        - ADV_gates: dictionary with however many gates we have to the region in xyz space; we can include the face tracer here
        - tracer: dictionary with the gate face tracer labeled
        - binmidTracer: the tracer midpoints from layers
        - nTm1: the length of the outputted information we want
        - binwidth: binwidthT1 or binwidthS1 which we want to normalize by

    outputs
        - gates: dictionary with the four gates in tracer space
    '''
    ADV = ADV_gates['ADV']
    TRACER = ADV_gates['TRACER']

    out = {}
    out['ADVh'] = {}
    out['G'] = {}
    out['ADVh']['sum'] = np.zeros((nTm1))
    out['G']['sum'] = np.zeros((nTm1))

    for gate in ADV.keys():
        # get the tracer
        tracer = TRACER[gate]
        tracer_flat = tracer.ravel()
        ADV_flat   = ADV[gate].ravel()

        # define the tracer values as bins
        tracer_flat = tracer.ravel()
        bin_idx_mid = np.digitize(tracer_flat, binmidTracer, right=False) - 1
        valid_mid   = (bin_idx_mid >= 0) & (bin_idx_mid < nTm1) & np.isfinite(tracer_flat)
        idx_mid     = bin_idx_mid[valid_mid]
        
        # get the per-bin sums
        ADVh_tracer = _bincount_sum_with_nan(idx_mid, ADV_flat[valid_mid], nTm1) # we should also check this is binning correctly
    
        # edge-based G (m^3/s): divide by edge binwidths
        G_tracer = ADVh_tracer / binwidth

        # set these in the dictionary under the keys
        out['ADVh'][gate] = ADVh_tracer
        out['G'][gate] = G_tracer

        # add this to the sum array
        out['ADVh']['sum'] += ADVh_tracer
        out['G']['sum'] += G_tracer
    
    return out # contains the information for however many keys we want
```

```python
G_out = gateway3D(ADV_gates,binmidT,nTm1,binwidthT1)
```
