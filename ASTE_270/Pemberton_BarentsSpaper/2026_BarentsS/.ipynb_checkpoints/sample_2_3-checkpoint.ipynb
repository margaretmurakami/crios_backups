```python
# we need DFh, DFr, ADVr

## do the advective convergence
file_name = "budg3d_hflux_set2"
meta_budg3d_hflux_set2 = parsemeta(dirdiags + file_name + "." + tsstr[0] + ".meta")
fldlist = np.array(meta_budg3d_hflux_set2["fldList"])
varnames = np.array(["DFxE_TH","DFyE_TH"])
recs = np.array([])
for var in varnames:
    irec = np.where(fldlist == var)
    recs = np.append(recs, irec[0][0])
DFxE_TH,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[0])
DFyE_TH,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[1])


# now 3d zfluxes
file_name = "budg3d_zflux_set1"
meta_budg3d_zflux_set1 = parsemeta(dirdiags + file_name + "." + tsstr[0] + ".meta")
fldlist = np.array(meta_budg3d_zflux_set1["fldList"])
varnames = np.array(["DFrE_TH","DFrI_TH"])
recs = np.array([])
for var in varnames:
    irec = np.where(fldlist == var)
    recs = np.append(recs, irec[0][0])
DFrE_TH,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[0])
DFrI_TH,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[1])
DFrE_TH = DFrE_TH.reshape(nz,ny,nx)
DFrI_TH = DFrI_TH.reshape(nz,ny,nx)

DF_hconv = calc_UV_conv_mod(nfx, nfy,get_aste_faces(DFxE_TH.reshape(nz, ny, nx), nfx, nfy),get_aste_faces(DFyE_TH.reshape(nz, ny, nx), nfx, nfy))
DF_hconv = DF_hconv * hf   # degCm^3/s at cell centers (matches: ff.DFh = ff.DFh .* hf)
DFhT = DF_hconv

trWtopDF = -(DFrE_TH+DFrI_TH)

DFrT = np.zeros((nz,ny,nx),dtype=float)
DFrT[:-1,:,:] = (trWtopDF[:-1] - trWtopDF[1:])

# reshape the DF terms
trWtopDF = trWtopDF.reshape((nz,ny,nx))
DFxE_TH = DFxE_TH.reshape((nz,ny,nx))
DFyE_TH = DFyE_TH.reshape((nz,ny,nx))

# now 3d zfluxes
file_name = "budg3d_zflux_set1"
meta_budg3d_zflux_set1 = parsemeta(dirdiags + file_name + "." + tsstr[0] + ".meta")
fldlist = np.array(meta_budg3d_zflux_set1["fldList"])
varnames = np.array(["ADVr_TH"])
recs = np.array([])
for var in varnames:
    irec = np.where(fldlist == var)
    recs = np.append(recs, irec[0][0])
ADVr_TH,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[0])
ADVr_TH = ADVr_TH.reshape(nz,ny,nx)
```

```python
# we still need to load the KPP and surface terms

# to get the surface term, we need J/s and convert to degC.m^3/s
file_name = 'budg2d_zflux_set1'
meta_budg2d_zflux_set1 = parsemeta(dirdiags + file_name + "." + tsstr[0] + ".meta")
fldlist = np.array(meta_budg2d_zflux_set1["fldList"])
varnames = np.array(["TFLUX","oceQsw","SItflux"])
recs = np.array([])
for var in varnames:
    irec = np.where(fldlist == var)
    recs = np.append(recs, irec[0][0])
TFLUX,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[0])
oceQsw,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[1])
SItflux,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[2])
TFLUX = TFLUX.reshape(ny,nx)
oceQsw = oceQsw.reshape(ny,nx)
SItflux = SItflux.reshape(ny,nx)

# we need to create zconv_top and swtop
dd = mygrid['RF'][:-1]
swfrac = 0.62*np.exp(dd/0.6)+(1-0.62)*np.exp(dd/20)
swfrac[dd < -200] = 0
swtop=mk3D_mod(swfrac,np.zeros((nz,ny,nx)))*mk3D_mod(RAC*oceQsw,np.zeros((nz,ny,nx)))   # J/s

# zconvtop_heat is here
zconv_top_heat = TFLUX * RAC     # W/m^2 * m^2 = J/s


def surface_contrib_JT(zconv_top_heat, swtop, rcp, fill_last=0.0):
    """
    zconv_top_heat: (ny, nx)
    swtop:          (nz, ny, nx)
    rcp:            scalar
    fill_last:      value for bottom slice (k = nz-1), usually 0.0 or np.nan
    returns:
      JsurfT:       (nz, ny, nx)  # Sv / PSU
    """
    nz, ny, nx = swtop.shape

    eT = zconv_top_heat.reshape(1, ny, nx)  # (1,ny,nx) for broadcast

    J = np.empty_like(swtop, dtype=float)

    # k = 0: (eT - fT[1]) / rcp / dT / dS * 1e-6
    J[0] = (eT[0] - swtop[1]) / rcp if np.ndim(binwidthT)==0 else \
           (eT[0] - swtop[1]) / rcp

    # 1 .. nz-2: -(fT[k+1]-fT[k]) / rcp / dT / dS * 1e-6
    J[1:nz-1] = -(swtop[2:nz] - swtop[1:nz-1]) / rcp

    # bottom slice (k = nz-1): no k+1; choose your boundary convention
    J[-1] = fill_last
    return J

Ft_surftest = surface_contrib_JT(zconv_top_heat,swtop,myparms['rcp'])    # this is in degC.m^3/s

# read kpp tend and from 3d zflux
file_name = "budg3d_kpptend_set1"
meta_budg3d_kpptend_set1 = parsemeta(dirdiags + file_name + "." + tsstr[0] + ".meta")
fldlist = np.array(meta_budg3d_kpptend_set1["fldList"])
varnames = np.array(["KPPg_TH"])
recs = np.array([])
for var in varnames:
    irec = np.where(fldlist == var)
    recs = np.append(recs, irec[0][0])
KPPg_TH,its,meta = rdmds(os.path.join(dirdiags, file_name),t2,returnmeta=True,rec=recs[0])
KPPg_TH = KPPg_TH.reshape(nz,ny,nx)

# do the vertical convergence for KPP
trWtopKPP = -(KPPg_TH)         # degC.m^3/s

tmpkpp = np.full((nz,ny,nx),np.nan)
tmpkpp[:-1,:,:] = trWtopKPP[:-1] - trWtopKPP[1:]

```

```python
# recreate a gradient-based T tend

def bin_cell_tmptend_to_theta(
    tmptend, THETA, iz, iy, ix,
    theta_edges,
    dx=1.0, dy=1.0, dz=1.0,
    Nsub=8,                  # Nsub^3 sub-samples per cell
):
    """
    Distribute ONE cell-centered tmptend[iz,iy,ix] into multiple THETA bins
    by reconstructing a linear THETA field inside the cell and sub-sampling.

    tmptend: (nz,ny,nx) cell-centered tendency already in degC*m^3/s (cell-integrated)
    THETA:   (nz,ny,nx) cell-centered temperature (degC)
    theta_edges: 1D bin edges (monotonic), length nbins+1
    dx,dy,dz: cell sizes (scalars here; adapt if you have arrays)
    """

    nz, ny, nx = THETA.shape
    Tc = THETA[iz, iy, ix]
    A  = tmptend[iz, iy, ix]   # what you want to accumulate in T-space

    # --- estimate face temperatures from neighboring cell centers (2nd-order-ish)
    # Use average between this cell and neighbor as a proxy for face-center temperature.
    # At boundaries, fall back to Tc.
    def face_temp(iz2, iy2, ix2):
        if (0 <= iz2 < nz) and (0 <= iy2 < ny) and (0 <= ix2 < nx):
            return 0.5 * (Tc + THETA[iz2, iy2, ix2])
        return Tc

    Txp = face_temp(iz, iy, ix+1)
    Txm = face_temp(iz, iy, ix-1)
    Typ = face_temp(iz, iy+1, ix)
    Tym = face_temp(iz, iy-1, ix)
    Tzp = face_temp(iz+1, iy, ix)
    Tzm = face_temp(iz-1, iy, ix)

    # --- gradients for linear reconstruction inside the cell
    gx = (Txp - Txm) / dx
    gy = (Typ - Tym) / dy
    gz = (Tzp - Tzm) / dz

    # --- sub-sample points inside cell (uniform grid of subcell centers)
    # cell is centered at (0,0,0) with extents +/- dx/2, etc.
    # subcell centers along each axis:
    s = (-0.5 + (np.arange(Nsub) + 0.5) / Nsub)
    xs = s * dx
    ys = s * dy
    zs = s * dz

    # Evaluate THETA at each sub-sample point:
    # T = Tc + gx*x + gy*y + gz*z
    # Use broadcasting to form (Nsub,Nsub,Nsub)
    Tsub = Tc + gx*xs[:,None,None] + gy*ys[None,:,None] + gz*zs[None,None,:]

    # Each sub-sample gets equal share of the cell-integrated tendency A
    w = A / (Nsub**3)

    # Bin: digitize gives indices 1..nbins; convert to 0..nbins-1
    b = np.digitize(Tsub.ravel(), theta_edges) - 1
    nbins = len(theta_edges) - 1

    # Keep only in-range bins
    m = (b >= 0) & (b < nbins)
    out = np.zeros(nbins, dtype=float)
    out += np.bincount(b[m], weights=np.full(np.count_nonzero(m), w), minlength=nbins)
    return Tsub,out
```

```python

def bin_cell_tmptend_to_theta(
    tmptend, THETA, iz, iy, ix,
    theta_edges,
    dx=1.0, dy=1.0, dz=1.0,
    Nsub=8,
):
    nz, ny, nx = THETA.shape

    Tc = THETA[iz, iy, ix]
    Ac = tmptend[iz, iy, ix]   # cell-integrated tendency (degC*m^3/s)

    def face_val(F, iz2, iy2, ix2, Fc, Tc=None):
        # second-order-ish face proxy: average with neighbor, fallback to center
        if (0 <= iz2 < nz) and (0 <= iy2 < ny) and (0 <= ix2 < nx):
            return 0.5 * (Fc + F[iz2, iy2, ix2])
        return Fc

    # ---- THETA reconstruction
    Txp = face_val(THETA, iz, iy, ix+1, Tc)
    Txm = face_val(THETA, iz, iy, ix-1, Tc)
    Typ = face_val(THETA, iz, iy+1, ix, Tc)
    Tym = face_val(THETA, iz, iy-1, ix, Tc)
    Tzp = face_val(THETA, iz+1, iy, ix, Tc)
    Tzm = face_val(THETA, iz-1, iy, ix, Tc)

    gxT = (Txp - Txm) / dx
    gyT = (Typ - Tym) / dy
    gzT = (Tzp - Tzm) / dz

    # ---- tmptend reconstruction
    Axp = face_val(tmptend, iz, iy, ix+1, Ac)
    Axm = face_val(tmptend, iz, iy, ix-1, Ac)
    Ayp = face_val(tmptend, iz, iy+1, ix, Ac)
    Aym = face_val(tmptend, iz, iy-1, ix, Ac)
    Azp = face_val(tmptend, iz+1, iy, ix, Ac)
    Azm = face_val(tmptend, iz-1, iy, ix, Ac)

    gxA = (Axp - Axm) / dx
    gyA = (Ayp - Aym) / dy
    gzA = (Azp - Azm) / dz

    # ---- subsample locations (cell-centered)
    s = (-0.5 + (np.arange(Nsub) + 0.5) / Nsub)
    xs = s * dx
    ys = s * dy
    zs = s * dz

    # ---- extrapolated subcell fields
    Tsub = Tc + gxT*xs[:,None,None] + gyT*ys[None,:,None] + gzT*zs[None,None,:]
    Asub = Ac + gxA*xs[:,None,None] + gyA*ys[None,:,None] + gzA*zs[None,None,:]

    # ---- bin by Tsub, weight by Asub (preserves sum if linear reconstruction)
    b = np.digitize(Tsub.ravel(), theta_edges) - 1
    nbins = len(theta_edges) - 1
    m = (b >= 0) & (b < nbins)

    w = (Asub.ravel()[m]) / (Nsub**3)   # each subsample represents equal volume
    out = np.bincount(b[m], weights=w, minlength=nbins).astype(float)

    return Tsub, Asub, out

Tsub1, Tthetadr_sub1, cell_hist1 = bin_cell_tmptend_to_theta(THETADR[0]/DRF3d, THETA, iz, iy, ix, binmidT, dx=DX, dy=DY, dz=DZ, Nsub=10)
Tsub2, Tthetadr_sub2, cell_hist2 = bin_cell_tmptend_to_theta(THETADR[1]/DRF3d, THETA, iz, iy, ix, binmidT, dx=DX, dy=DY, dz=DZ, Nsub=10)
```

```python
def bin_cell_tmptend_to_theta_return_subcell(
    tmptend, THETA, iz, iy, ix,
    theta_edges,
    dx=1.0, dy=1.0, dz=1.0,
    Nsub=8,
):
    """
    tmptend: (nz,ny,nx) cell-integrated tendency (degC*m^3/s) at cell centers
    Returns:
      Tsub:     (Nsub,Nsub,Nsub) reconstructed temperatures (degC)
      Asub_int: (Nsub,Nsub,Nsub) subcell-integrated contributions (degC*m^3/s)
      hist:     (nbins,) binned sum (degC*m^3/s) over theta bins
      edges:    (Nsub+1,) coordinate edges for each axis (in physical units)
    """
    nz, ny, nx = THETA.shape
    Tc = THETA[iz, iy, ix]
    A  = tmptend[iz, iy, ix]  # degC*m^3/s for the whole cell

    def face_temp(iz2, iy2, ix2):
        if (0 <= iz2 < nz) and (0 <= iy2 < ny) and (0 <= ix2 < nx):
            return 0.5 * (Tc + THETA[iz2, iy2, ix2])
        return Tc

    Txp = face_temp(iz, iy, ix+1); Txm = face_temp(iz, iy, ix-1)
    Typ = face_temp(iz, iy+1, ix); Tym = face_temp(iz, iy-1, ix)
    Tzp = face_temp(iz+1, iy, ix); Tzm = face_temp(iz-1, iy, ix)

    gx = (Txp - Txm) / dx
    gy = (Typ - Tym) / dy
    gz = (Tzp - Tzm) / dz

    # subcell centers
    s  = (-0.5 + (np.arange(Nsub) + 0.5) / Nsub)
    xs = s * dx
    ys = s * dy
    zs = s * dz

    Tsub = Tc + gx*xs[:,None,None] + gy*ys[None,:,None] + gz*zs[None,None,:]

    # subcell-integrated tendency contributions (uniform split; sums to A)
    Asub_int = np.full((Nsub, Nsub, Nsub), A/(Nsub**3), dtype=float)

    # bin by Tsub using per-subcell contributions
    b = np.digitize(Tsub.ravel(), theta_edges) - 1
    nbins = len(theta_edges) - 1
    m = (b >= 0) & (b < nbins)

    hist = np.bincount(b[m], weights=Asub_int.ravel()[m], minlength=nbins).astype(float)

    # physical coordinate edges along each axis (centered cell)
    x_edges = (np.linspace(-0.5, 0.5, Nsub+1) * dx)
    y_edges = (np.linspace(-0.5, 0.5, Nsub+1) * dy)
    z_edges = (np.linspace(-0.5, 0.5, Nsub+1) * dz)

    return Tsub, Asub_int, hist, (x_edges, y_edges, z_edges)

Tsub, Asub_int, cell_hist, (x_edges, y_edges, z_edges) = bin_cell_tmptend_to_theta_return_subcell(
    tmptend, THETA, iz, iy, ix,
    theta_edges=binmidT,
    dx=DX, dy=DY, dz=DZ,
    Nsub=100
)

print("check sum:", Asub_int.sum(), "vs A:", tmptend[iz,iy,ix])

Tsub, vol_sub_int, cell_hist, (x_edges, y_edges, z_edges) = bin_cell_tmptend_to_theta_return_subcell(
    vol, THETA, iz, iy, ix,
    theta_edges=binmidT,
    dx=DX, dy=DY, dz=DZ,
    Nsub=100
)

print("check sum:", vol_sub_int.sum(), "vs A:", vol[iz,iy,ix])
```
