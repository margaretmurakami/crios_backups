# show the seasonality of the release of these polynyas

```python
# we need to move nan to end for the selected columns and plot the Xgrid and Ygrid of the first release point
def move_nans_to_end(Xgrid):
    # Get shape
    m, n = Xgrid.shape  
    result = np.full_like(Xgrid, np.nan)  # Initialize result with NaNs

    for col in range(n):
        non_nan_values = Xgrid[:, col][~np.isnan(Xgrid[:, col])]  # Extract non-NaN values
        result[:len(non_nan_values), col] = non_nan_values  # Place them at the top

    return result
```

```python
Xgrid = ds.Xgrid.values[:,combined_columns]
Ygrid = ds.Ygrid.values[:,combined_columns]
depth = ds.depth.values[:,combined_columns]
temp = ds.temp.values[:,combined_columns]
salt = ds.salt.values[:,combined_columns]
rho = ds.rho.values[:,combined_columns]
wmt_subset = wmt[:,combined_columns]
```

```python
Xgrid_rearranged = move_nans_to_end(Xgrid)
Ygrid_rearranged = move_nans_to_end(Ygrid)
rho_rearranged = move_nans_to_end(rho)
print('a')
temp_rearranged = move_nans_to_end(temp)
salt_rearranged = move_nans_to_end(salt)
print('b')
depth_rearranged = move_nans_to_end(depth)
wmt_rearranged = move_nans_to_end(wmt_subset)
```

```python
# for the last row of wmt
thismass = wmt_rearranged[35040]

# get counts
unique_elements, counts_elements = np.unique(thismass, return_counts=True)

# ensure all water masses (0-7) are represented
for i in range(8):
    if not np.any(unique_elements == i):
        unique_elements = np.append(unique_elements, i)
        counts_elements = np.append(counts_elements, 0)

# sort to align water masses correctly
sort_idx = np.argsort(unique_elements)
unique_elements = unique_elements[sort_idx]
counts_elements = counts_elements[sort_idx]

# key for water masses
water_mass_labels = [
    'AABW',   # 0
    'mCDW',   # 1
    'ISW',    # 2
    'DSW',    # 3
    'AASW',   # 4
    'WW',     # 5
    'MSW',    # 6
    'nan/beached'  # 7
]

# print counts for each water mass
for label, count in zip(water_mass_labels, counts_elements):
    print(f"{label}: {count}")
```

```python
import numpy as np
import netCDF4 as nc

# Example shapes
# ocean_time dimension (35040)
# drifter dimension (number of columns, e.g., n_drifters)

# Let’s assume you’ve already run:
# Xgrid_rearranged = move_nans_to_end(Xgrid)
# and similarly for Ygrid, rho, temp, salt, depth, wmt_subset.

# Slice up to row 35040
max_rows = 35040
Xgrid_rearranged = Xgrid_rearranged[:max_rows, :]
Ygrid_rearranged = Ygrid_rearranged[:max_rows, :]
rho_rearranged = rho_rearranged[:max_rows, :]
temp_rearranged = temp_rearranged[:max_rows, :]
salt_rearranged = salt_rearranged[:max_rows, :]
depth_rearranged = depth_rearranged[:max_rows, :]
wmt_rearranged = wmt_rearranged[:max_rows, :]

# Determine number of drifters
n_drifters = Xgrid_rearranged.shape[1]

# Create a new NetCDF file
with nc.Dataset('/scratch/mmurakami/WAOM/rearranged_output.nc', 'w', format='NETCDF4') as ds:
    # Define dimensions
    ds.createDimension('ocean_time', max_rows)
    ds.createDimension('drifter', n_drifters)

    # Create variables
    x_var = ds.createVariable('Xgrid', 'f4', ('ocean_time', 'drifter'))
    y_var = ds.createVariable('Ygrid', 'f4', ('ocean_time', 'drifter'))
    rho_var = ds.createVariable('rho', 'f4', ('ocean_time', 'drifter'))
    temp_var = ds.createVariable('temp', 'f4', ('ocean_time', 'drifter'))
    salt_var = ds.createVariable('salt', 'f4', ('ocean_time', 'drifter'))
    depth_var = ds.createVariable('depth', 'f4', ('ocean_time', 'drifter'))
    wmt_var = ds.createVariable('wmt', 'f4', ('ocean_time', 'drifter'))

    # Add some global attributes (optional)
    ds.description = 'Rearranged dataset with NaNs moved to end and sliced to row 35040.'
    ds.source = 'Created by move_nans_to_end rearrangement script.'

    # Fill variables
    x_var[:, :] = Xgrid_rearranged
    y_var[:, :] = Ygrid_rearranged
    rho_var[:, :] = rho_rearranged
    temp_var[:, :] = temp_rearranged
    salt_var[:, :] = salt_rearranged
    depth_var[:, :] = depth_rearranged
    wmt_var[:, :] = wmt_rearranged

print("✅ New NetCDF file 'rearranged_output.nc' saved successfully!")
```

```python
# label the points based on where they fall
start_x = dstest.Xgrid.values[0]  # shape (4458,)
start_y = dstest.Ygrid.values[0]
```

```python
from shapely.geometry import Point

# Create Points
start_points = [Point(x, y) for x, y in zip(start_x, start_y)]

# List of polygons and labels
polygons = [darnley, mackenzie, davis, barrier]
labels = ["darnley", "mackenzie", "davis", "barrier"]

# Initialize labels array
drifter_labels = np.full(len(start_points), "other", dtype='<U20')

# Assign labels
for i, pt in enumerate(start_points):
    for poly, label in zip(polygons, labels):
        if poly.contains(pt):
            drifter_labels[i] = label
            break

# Now drifter_labels[i] contains the polygon name (or "other") for each drifter!
print(np.unique(drifter_labels, return_counts=True))
```

```python
with nc.Dataset('drifter_labels.nc', 'w', format='NETCDF4') as ds:
    ds.createDimension('drifter', len(drifter_labels))
    label_var = ds.createVariable('drifter_label', str, ('drifter',))
    label_var[:] = drifter_labels
```
